# Weather Data Visualization using Python

```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

* Making variable and taking dataset

```python
weather = pd.read_csv('Test.csv')
```

* Cheking the heading of the data

```python
weather.head()
```

* Checking all the data types and data values

```python
weather.info()
```

# Relationship Between Humidity and Temperature








```python
sns.barplot(x=weather['humidity'], y=weather['temperature'], palette="viridis")
```

```python
import matplotlib.pyplot as plt
plt.show()
```

# Distribution of Humidity Levels

```python
sns.distplot(weather['humidity'])
```

# Joint Distribution of Humidity and Temperature

```python
sns.jointplot(x=weather['humidity'], y=weather['temperature'])       
```

```python
sns.jointplot(x=weather['humidity'], y=weather['temperature'], kind='hex')
```

```python
sns.jointplot(x=weather['humidity'], y=weather['temperature'], kind='kde')
```

#  Pairwise Relationships (Pair Plot)

```python
sns.pairplot(weather[['humidity', 'temperature', 'air_pollution_index']])
```

# Weather Type vs Temperature (Strip Plot)

```python
sns.stripplot(x=weather['weather_type'], y=weather['temperature'],palette="viridis")
```

* Jitter prevents overlapping by spreading points for better visibility.








```python
sns.stripplot(x=weather['weather_type'], y=weather['temperature'],jitter=True,palette="viridis")
```

# Temperature vs Humidity by Weather Type (Box Plot)

```python
sns.boxplot(x=weather['humidity'], y=weather['temperature'], hue=weather['weather_type'])
```

# Temperature vs Humidity by Weather Type (Bar Plot)

```python
sns.barplot(x=weather['humidity'],y=weather['temperature'],hue=weather['weather_type'],palette="viridis")
```

# Frequency of Weather Types (Count Plot)

```python
sns.countplot(x=weather['weather_type'],palette="viridis")
```

# Weather Point Plot








```python
sns.pointplot(x=weather['humidity'],y=weather['temperature'],hue=weather['weather_type'],palette="viridis")
```

# Humidity-Temperature Regression Plot

```python
sns.lmplot(x='humidity',y='temperature',data=weather)
```

* The plot shows the linear relationship between humidity and temperature.

# Weather-Dependent Humidity-Temperature Regression Plot

```python
sns.lmplot(x="humidity", y="temperature", hue="weather_type", data=weather, palette="viridis")
```
